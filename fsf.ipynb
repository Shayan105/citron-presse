```python
import pickle
```

```python
#open daily_topics.pickle

with open("daily_themes.pkl", "rb") as f:
    daily_topics = pickle.load(f)
# print the topics
dict2 = {}
for key, value in daily_topics.items():
    print(f"{key} : {value}")
    key2 = key.replace("G:/", "")
    dict2[key2] = value
    
```

```python
print(dict2)
# save the topics in a file named daily_themes.json
with open("daily_themes.pkl", "w") as f:
    import json
    json.dump(dict2, f)  # ensure_ascii=False to keep non-ASCII characters

```
